```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Churn'].value_counts()
```

```python
sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()
```

Dataset is imbalanced
Churn class is minority

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title("Tenure vs Churn")
plt.show()
```

Customers with short tenure show a much higher likelihood of churn.

As tenure increases, churn probability drops sharply.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

Churned customers generally have higher monthly charges than retained customers.
Customers may feel they are overpaying relative to value received.

```python
numeric_df = df[['SeniorCitizen', 'tenure', 'MonthlyCharges']]
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
